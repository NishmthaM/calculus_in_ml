```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

**1.Coin toss simulation**

```python
n=500
outcome=np.random.choice(["H","T"],size=n)
print(outcome[:5])
print(len(outcome))
```

```python
head=np.sum(outcome=="H")
tail=np.sum(outcome=="T")

print("Head:",head)
print("Tail:",tail)
print("Estimated Probablity of heads:",head/n)
print("Estimated Probablity of tails:",tail/n)
```

**2.Dice roll simulation**

```python
n=1000

dice= np.random.randint(1,7, size=n)

p_four = np.mean(dice==4)

print("Estimated P(4):",p_four)
vals,counts = np.unique(dice, return_counts=True)

print(vals)
print(counts)
```
