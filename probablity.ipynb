```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
n=500
outcome=np.random.choice(["H","T"],size=n)
print(outcome[:5])
print(len(outcome))
```

```python
head=np.sum(outcome=="H")
tail=np.sum(outcome=="T")

print("Head:",head)
print("Tail:",tail)
print("Estimated Probablity of heads:",head/n)
print("Estimated Probablity of tails:",tail/n)
```
