```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True)
```

**1.A simple line: y=2x+1**

```python
def line_fun(x):
  return 2*x+1

```

```python
x=np.linspace(-3, 3, 100)
y=line_fun(x)
```

```python
print(x[:5])
print(y[:5])
```

```python
plt.figure(figsize=(6,3))
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(alpha=0.3)
plt.show()

x1 , x2= -2,1
y1,y2=line_fun(x1),line_fun(x2)

rise=y2-y1
run=x2-x1

slope=rise/run
print()
print("Line Example")
print()
print("Point 1:",f"({x1},{y1})")
print("Point 2:",f"({x2},{y2})")
print("Rise:",rise)
print("Run:",run)
print("Slope:",slope)
```

**Curve : y=x^2**

```python
def curve(x):
  return x**2

x_val = np.linspace(-3, 3, 200)
y_val = curve(x_val)

plt.figure(figsize=(6,4))
plt.plot(x_val,y_val)
plt.title("Curve: y^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(alpha=0.3)
plt.show()
```

```python
def slope_est(func, x1, h=0.00001):
  x2= x1+h
  y1= curve(x1)
  y2= func(x2)
  dy= y2-y1
  dx= h
  slope= dy/dx
  return slope
```

```python
points =[-2, 0, 2]
print("Slope estimates")

for p in points:
  s= slope_est(curve, p)
  print(f"x={p} estimated slope = {s:.2f}")
```
